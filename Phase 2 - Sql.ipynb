##### Using regex function for counting no of single beds and double beds

- double, extra-large double, large double are considered as double beds
- futon, sofa, bunk, single are considered as single beds

```python
import re



# creating empty lists for single bed and double bed counts
single_bed_counts = []
double_bed_counts = []

# regex pattern to match various types of beds
bed_pattern = r'(\d+)\s*(futon|sofa|bunk|single|double|extra-large double|large double)\s*bed'

# Iterating through the bed data and extract bed counts for each hotel
for hotel in bed_data:
    bed_matches = re.findall(bed_pattern, hotel)
 
    hotel_single_bed_count = 0
    hotel_double_bed_count = 0

    for count, bed_type in bed_matches:
        count = int(count)
        if bed_type in ["double", "extra-large double", "large double"]:
            hotel_double_bed_count += count
        else:
            hotel_single_bed_count += count

    single_bed_counts.append(hotel_single_bed_count)
    double_bed_counts.append(hotel_double_bed_count)

print(single_bed_counts,'\n',double_bed_counts)
```

```python
data3 = pd.DataFrame({ 'sno':sl_no ,  'Hotel name':name_data, 'Room Type':room_type_data ,'Single Bed':single_bed_counts ,"Double bed":double_bed_counts  ,'Prices':prices_data})
data3.sample(3)
```
