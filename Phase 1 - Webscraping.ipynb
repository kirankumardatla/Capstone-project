```python
# clicking x button of popup if popup found
time.sleep(5)
try:
    pop_btn = WebDriverWait(driver,20).until(EC.presence_of_element_located((By.XPATH,'//button[@aria-label="Dismiss sign in information."]')))
    pop_btn.click()
except:
    pass    


# clocking cities tab

cities_tab = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'/html/body/div[5]/div[2]/nav/ul/li[2]/button')))
cities_tab.location_once_scrolled_into_view
cities_tab.click()
```

```python
#Scraping Cities url
Series_data = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//li[@class="bui-carousel__item"]//div[@class="wcu-seo-popular-city bui-spacer--large"]//a')))

cities = []
c_urls = [] 
for i in Series_data:
    if i.text.split("\n")[0] in ['Dubai','London','Kuala Lumpur','Manchester','New Delhi','Birmingham','Berlin','Sydney','Melbourne','Paris','Tokyo','Toronto']:
        cities.append(i.text.split("\n")[0])
        c_urls.append(i.get_attribute('href'))
```

```python
# creating empty list for storing hotel urls
h_url = []

#Scraping Hotel urls
for c_url in c_urls:

    driver.get(c_url)

# selecting random dates for getting price data    
    date_start = cities_tab = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//td[@data-date="2023-10-29"]')))
    date_start.click()
    time.sleep(0.5)

    date_start = cities_tab = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//td[@data-date="2023-10-30"]')))
    date_start.click()
    time.sleep(0.5)

    search_btn = WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,'//button[@class="sb-searchbox__button "]')))
    search_btn.click()
    time.sleep(0.5)
    while True:

        try:
            hotel_data  = WebDriverWait(driver,10).until(EC.presence_of_all_elements_located((By.XPATH,'//a[@class="e13098a59f"]')))
            for i in hotel_data:
                h_url.append(i.get_attribute('href'))

            nxt_btn  = WebDriverWait(driver,5).until(EC.presence_of_element_located((By.XPATH,'//div[@class="b16a89683f cab1524053"]//button[@aria-label="Next page"]')))
            nxt_btn.location_once_scrolled_into_view
            nxt_btn.click()
            time.sleep(5)
        except:
            break
print(len(h_url))    
    
    
```

```python
# creating variables for all xpaths
name_xp = '//h2[@class="d2fee87262 pp-header__title"]'
city_xp = '//input[@name="ss"]'
loc_xp = '//p[@class="address address_clean"]/span[1]'
review_xp = '//div[@data-testid="review-score-component"]'
star_r_xp = '//span[@class="fcd9eec8fb d31eda6efc c25361c37f"]'
brekfast_c_xp = '//span[@class="ph-item-copy-breakfast-option"]'
facilities_xp = '//span[@class="db312485ba"]'
room_type_xp = '//div[@class="hprt-roomtype-block hprt-roomtype-name hp_rt_room_name_icon__container"]'
bed_det_xp = '//div[@class="hprt-roomtype-bed"]'
prices_xp = '//div[@class="bui-price-display__value prco-text-nowrap-helper prco-inline-block-maker-helper prco-f-font-heading "]'
```

```python
# creating empty lists for storing data
name_data = []
city_data = []
loc_data = []
no_of_rev_data = []
user_rating_data = []
rev_score_data = []
star_rating_data = []
wifi_data = []
fam_room_data = []
smok_data = []
restaurent_data = []
Bar_data = []
heating_data = []
Lift_data = []
Breakfast_c1_data = []
Breakfast_c2_data = []
Breakfast_c3_data = []
room_type_data = []
bed_data = []
prices_data = []
```

webscraping data to respective lists

```python
# iterating over all hotel urls
for i in h_url:
    while True:
        try:
            driver.get(i)
            break
        except:
            time.sleep(30)
            pass 
        
    # Hotel Name data    
    try:
        name = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,name_xp)))
        name_data.append(name.text)
    except:
        name_data.append('null')    

    # City Name data    
    try:
        city = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,city_xp)))
        city_data.append(city.get_attribute('value'))
    except:
        city_data.append('null')

    # Location Data        
    try:
        loc = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,loc_xp)))
        loc_data.append(loc.text)
    except:
        loc_data.append('null')

    # Reviews and Ratings Data    
    try:
        rev = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,review_xp)))
        user_rating_data.append(rev.text.split('\n')[0])
        rev_score_data.append(rev.text.split('\n')[1].split('·')[0][:-1])
        no_of_rev_data.append(rev.text.split('\n')[1].split('·')[1].split()[0])
    except:
        user_rating_data.append('null')
        rev_score_data.append('null')
        no_of_rev_data.append('null')   

    # Star rating Data    
    try:
        star_r = WebDriverWait(driver,2).until(EC.presence_of_all_elements_located((By.XPATH,star_r_xp)))
        star_rating_data.append(len(star_r))
    except:
        star_rating_data.append('null')    



    # Facilities Data
    try:

        facilities  = WebDriverWait(driver,3).until(EC.presence_of_all_elements_located((By.XPATH,facilities_xp)))

        facilities_all = []
        for i in facilities:
            facilities_all.append(i.text)
    except:
        facilities_all = ['null','null']


    if 'Free WiFi' in facilities_all:
        wifi_data.append('yes')
    else :
        wifi_data.append('no')

    if 'Family rooms' in facilities_all:
        fam_room_data.append('yes')
    else :
        fam_room_data.append('no')

    if ('Non-smoking rooms' or 'Non-smoking throughout') in facilities_all:
        smok_data.append('yes')
    else :
        smok_data.append('no')

    if 'Restaurant' in facilities_all:
        restaurent_data.append('yes')
    else :
        restaurent_data.append('no')

    if 'Bar' in facilities_all:
        Bar_data.append('yes')
    else :
        Bar_data.append('no')

    if 'Heating' in facilities_all:
        heating_data.append('yes')
    else :
        heating_data.append('no')

    if 'Lift' in facilities_all:
        Lift_data.append('yes')
    else :
        Lift_data.append('no')


    # Breakfast Data    
    try:
        brekfast_c  = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,brekfast_c_xp)))
        Breakfast_c1_data.append(brekfast_c.text.split(',')[0])
        try:
            Breakfast_c2_data.append(brekfast_c.text.split(',')[1])
        except:
            Breakfast_c2_data.append('null')    
        try:
            Breakfast_c3_data.append(brekfast_c.text.split(',')[2])
        except:
            Breakfast_c3_data.append('null')    


    except:
        Breakfast_c1_data.append('null')
        Breakfast_c2_data.append('null')
        Breakfast_c3_data.append('null')



        
    # Room Type Data
    try:
        room_type  = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,room_type_xp)))
        room_type_data.append(room_type.text)
    except:
        room_type_data.append('null')



        

    # Single or double bed data
    try:
        bedtyp  = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,bed_det_xp)))
        bed_data.append(bedtyp.text)
    except:
        bed_data.append('null')



        
    # Prices Data
    try:
        prices  = WebDriverWait(driver,3).until(EC.presence_of_element_located((By.XPATH,prices_xp)))
        prices_data.append(prices.text[2:])
    except:
        prices_data.append('null')   

    all_lists = [name_data,city_data,loc_data,user_rating_data,no_of_rev_data,rev_score_data,star_rating_data,wifi_data,fam_room_data,smok_data,restaurent_data,Bar_data,heating_data,Lift_data,Breakfast_c1_data,Breakfast_c2_data,Breakfast_c3_data,room_type_data,bed_data,prices_data] 

    # Calculate the minimum length among all the lists
    min_length = min(len(lst) for lst in all_lists)

    # Remove the last item from lists that have a length greater than the minimum length
    for lst in all_lists:
        if len(lst) > min_length:
            lst.pop()
         
```
